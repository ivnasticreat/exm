## Задание 1
Исследуйте систему векторов на линейную зависимость или независимость без использования Python (вручную):
а1 = (1, 8, -1)
а2 =  (-2, 3, 3)
a3 = (4, -11, 9)  

**Определение линейной зависимости**:

Векторы $a_1, a_2, a_3$ линейно зависимы, если существуют такие коэффициенты $c_1, c_2, c_3$, не все из которых равны нулю, что выполняется:

$
c_1 a_1 + c_2 a_2 + c_3 a_3 = 0.
$

Если таких коэффициентов нет, то векторы линейно независимы.

**Составим уравнения:**

Подставим векторы $a_1, a_2, a_3$:

$
c_1 (1, 8, -1) + c_2 (-2, 3, 3) + c_3 (4, -11, 9) = (0, 0, 0).
$

Система линейных уравнений:

$
\begin{aligned}
& \quad c_1 - 2c_2 + 4c_3 = 0, \\
& \quad 8c_1 + 3c_2 - 11c_3 = 0, \\
& \quad -c_1 + 3c_2 + 9c_3 = 0.
\end{aligned}
$


**Представим систему в матричном виде:**

$$
\begin{bmatrix}
1 & -2 & 4 \\
8 & 3 & -11 \\
-1 & 3 & 9
\end{bmatrix}
\cdot 
\begin{bmatrix}
c_1 \\
c_2 \\
c_3
\end{bmatrix}
=
\begin{bmatrix}
0 \\
0 \\
0
\end{bmatrix}
$$

Чтобы определить, существуют ли нетривиальные решения ($c_1, c_2, c_3$ не равные нулю), нужно найти ранг матрицы

**Находим ранг матрицы**
Найдем определитель:

$
\text{det}
\begin{bmatrix}
1 & -2 & 4 \\
8 & 3 & -11 \\
-1 & 3 & 9
\end{bmatrix}
$


$$
\text{det} =
1 \cdot
\begin{vmatrix}
3 & -11 \\
3 & 9
\end{vmatrix}
-
(-2) \cdot
\begin{vmatrix}
8 & -11 \\
-1 & 9
\end{vmatrix}
+
4 \cdot
\begin{vmatrix}
8 & 3 \\
-1 & 3
\end{vmatrix}
$$

$
\text{det} = 1 \cdot 60 - (-2) \cdot 61 + 4 \cdot 27 = 60 + 122 + 108 = 290
$


Так как определитель матрицы $ \neq 0 $, ранг матрицы равен 3 (максимальный). Следовательно, система уравнений имеет только тривиальное решение $c_1 = c_2 = c_3 = 0$


***Вывод:***

Векторы $a_1 = (1, 8, -1)$, $a_2 = (-2, 3, 3)$, $a_3 = (4, -11, 9)$ линейно независимы.  Это означает, что ни один из векторов нельзя представить как линейную комбинацию других векторов системы.

## Задание 10
Разделите данные на предикторы и целевую переменную. Обучите модель с помощью классической линейной регрессии. Линейную регрессию необходимо реализовать без готовых библиотек.

Оцените качество модели с помощью хотя бы двух метрик качества.

Распишите решение с помощью кода на языке Python и напишите текстом интерпретацию полученного вывода.

```python
import numpy as np

# Разделяем данные на предикторы и целевую переменную
X = numeric_cleaned.drop(['cnt', 'mnth'], axis=1)
y = numeric_cleaned['cnt']

# Добавляем столбец единиц для свободного члена
X = np.column_stack([np.ones(len(X)), X])

# Реализация линейной регрессии
class LinearRegression:
    def fit(self, X, y):
        self.coefficients = np.linalg.inv(X.T @ X) @ X.T @ y
        return self

    def predict(self, X):
        return X @ self.coefficients

# Обучаем модель
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Метрики качества
def r2_score(y_true, y_pred):
    ss_res = np.sum((y_true - y_pred) ** 2)
    ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)
    return 1 - (ss_res / ss_tot)

def mse(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

# Вычисляем метрики
r2 = r2_score(y, y_pred)
mse_value = mse(y, y_pred)

print("Метрики качества модели:")
print(f"R² score: {r2:.3f}")
print(f"MSE: {mse_value:.3f}")

```
